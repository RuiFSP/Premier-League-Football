Recommended Feature Engineering Ideas

    1. Goal Difference and Outcome Indicators:
        - GoalDifference: Compute the difference between FTHG and FTAG.
        - HomeWin, AwayWin, Draw: Encode FTR as binary indicators for each possible outcome.

    2. Aggregated Match Statistics:
        - TotalShots, TotalShotsOnTarget, TotalFouls, TotalCorners: Sum home and away statistics (e.g., HS + AS) to represent the overall game dynamics.
        - HomeShotAccuracy, AwayShotAccuracy: Calculate shot accuracy for home and away teams (e.g., HST/HS for home).

    3. Performance Metrics Based on Past Matches:
        - Moving Averages: For each team, calculate rolling averages over recent games for goals scored, goals conceded, fouls, yellow/red cards, etc. These can help reflect recent form.
        - Last 5 Match Results: Encode recent form by recording the outcomes of each team's last 5 games, e.g., 3 wins, 1 loss, and 1 draw.
    
    4. Time-Based Features:
        - Day of the Week: Extract the day of the week from Date to check if weekday/weekend affects results.
        - Month or Seasonality: Check if there are seasonal performance trends by encoding month or even period within the season (early, mid, late season).

    5. Relative Strength Indicators:
        - Home Advantage Factors: Calculate home statistics divided by away statistics for specific metrics like HS/AS, HST/AST, and HF/AF.
        - Expected Result: Using the betting odds (B365H, B365D, B365A), you could calculate implied probabilities for each outcome and compare them against actual results to capture any potential upsets.

    6. Additional Odds-Based Features:

        - Implied Win Probability: Convert betting odds to probabilities (e.g., Prob(HomeÂ Win)
        - $\text{Prob(Home Win)} = \frac{1}{B365H}$ to indicate how expected a win or loss was according to odds.

```python
# 1. Goal Difference and Outcome Indicators
df['GoalDifference'] = df['FTHG'] - df['FTAG']
df['HomeWin'] = (df['FTR'] == 'H').astype(int)
df['AwayWin'] = (df['FTR'] == 'A').astype(int)
df['Draw'] = (df['FTR'] == 'D').astype(int)

# 2. Aggregated Match Statistics
df['TotalShots'] = df['HS'] + df['AS']
df['TotalShotsOnTarget'] = df['HST'] + df['AST']
df['TotalFouls'] = df['HF'] + df['AF']
df['TotalCorners'] = df['HC'] + df['AC']
df['HomeShotAccuracy'] = df['HST'] / df['HS'].replace(0, 1)  # Avoid division by zero
df['AwayShotAccuracy'] = df['AST'] / df['AS'].replace(0, 1)

# 3. Time-Based Features
# Step 1: Create a copy of the original 'Date' column to preserve the original values
df['OriginalDate'] = df['Date']

# Step 2: Try parsing the 'Date' column with the '%d/%m/%y' format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y', errors='coerce')

# Step 3: For any NaT values, try parsing again with the '%d/%m/%Y' format
# Only apply this to rows where the 'Date' is NaT (i.e., failed first parsing)
df['Date'] = df['Date'].combine_first(pd.to_datetime(df['OriginalDate'], format='%d/%m/%Y', errors='coerce'))

# Step 4: Now that we have successfully parsed the dates, we can reformat them to 'dd/mm/yy'
df['Date'] = df['Date'].dt.strftime('%d/%m/%y')

#convert to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')

# Drop the 'OriginalDate' column as it's no longer needed
df.drop(columns=['OriginalDate'], inplace=True)

df['DayOfWeek'] = df['Date'].dt.dayofweek  # Monday=0, Sunday=6
df['Month'] = df['Date'].dt.month


# 4. Relative Strength Indicators
df['HomeShotsToAwayShots'] = df['HS'] / df['AS'].replace(0, 1)
df['HomeFoulsToAwayFouls'] = df['HF'] / df['AF'].replace(0, 1)

# 5. Betting Odds-Based Features
df['ImpliedHomeWinProb'] = 1 / df['B365H']
df['ImpliedDrawProb'] = 1 / df['B365D']
df['ImpliedAwayWinProb'] = 1 / df['B365A']

# Normalize probabilities to sum up to 1
total_prob = df['ImpliedHomeWinProb'] + df['ImpliedDrawProb'] + df['ImpliedAwayWinProb']
df['ImpliedHomeWinProb'] /= total_prob
df['ImpliedDrawProb'] /= total_prob
df['ImpliedAwayWinProb'] /= total_prob

# Show the newly created features
df[['GoalDifference', 'HomeWin', 'AwayWin', 'Draw', 'TotalShots', 'TotalShotsOnTarget',
      'TotalFouls', 'TotalCorners', 'HomeShotAccuracy', 'AwayShotAccuracy', 'DayOfWeek',
      'Month', 'HomeShotsToAwayShots', 'HomeFoulsToAwayFouls',
      'ImpliedHomeWinProb', 'ImpliedDrawProb', 'ImpliedAwayWinProb']].head()
```
