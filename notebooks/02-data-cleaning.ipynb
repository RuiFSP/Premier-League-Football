# Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
pd.set_option('display.max_columns',None)
```

```python
data = os.path.join('..','data', 'all_concat_football_data.csv')
data
```

```python
df = pd.read_csv(data)
df.head(30)
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.columns
```

```python
df.isna().sum()
```

```python
# check row with NaN
df[df.isnull().any(axis=1)]
```

```python
# convert date to datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')

# calculate the moving average of the last 5 games for each team
df['moving_average_3_games_home'] = df.groupby('hometeam')['fthg'].transform(lambda x: x.rolling(3, min_periods=3).mean())
df['moving_average_3_games_away'] = df.groupby('awayteam')['ftag'].transform(lambda x: x.rolling(3, min_periods=3).mean())

# calculate the moving average of the last 5 games for each team
df['moving_average_5_games_home'] = df.groupby('hometeam')['fthg'].transform(lambda x: x.rolling(5, min_periods=5).mean())
df['moving_average_5_games_away'] = df.groupby('awayteam')['ftag'].transform(lambda x: x.rolling(5, min_periods=5).mean())

# calculate the goal difference for each team
df['goal_difference_home'] = df['fthg'] - df['ftag']
df['goal_difference_away'] = df['ftag'] - df['fthg']
```

```python
df.head()
```

```python
# plot goal difference distribution for home and away teams
plt.figure(figsize=(12,6))
sns.histplot(df['goal_difference_home'], kde=True, color='blue', label='Home Goal Difference', alpha=0.5)
sns.histplot(df['goal_difference_away'], kde=True, color='red', label='Away Goal Difference', alpha=0.5)
plt.xlabel('Goal Difference')
plt.ylabel('Frequency')
plt.title('Goal Difference Distribution')
plt.legend()
plt.show()
```
